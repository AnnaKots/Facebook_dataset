## Анализ датасета Facebook

```python
import pandas as pd
import numpy as np
from IPython.display import HTML, display
import tabulate

data=pd.read_csv('C:/Users/kotsi/Downloads/dataset_Facebook.csv', sep=';')
data.head()
```

```python
# Для этих стобцов нам бесполезно считать среднее, максимальное и минимальное
uncounted_columns=data[['Type','Category','Post Month', 'Post Weekday', 'Post Hour', 'Paid']] #мода, медиана

numerical_columns=data.drop(['Category', 'Post Month', 'Post Weekday', 'Post Hour', 'Paid'], axis=1)

def static_uncounted(columns):
    measures=dict()
    measures['median']=columns.median()
    measures=pd.DataFrame(measures).T
    return measures

def static_counted(columns):
    measures=dict()
    measures['0_mean']=columns.mean()
    measures['1_max']=columns.max()
    measures['2_min']=columns.min()
    measures['3_median']=columns.median()
    measures=pd.DataFrame(measures).T
    return measures

static_uncounted(uncounted_columns)
```

```python
un_photos=uncounted_columns.drop(np.where(uncounted_columns['Type']!='Photo')[0])
photos=numerical_columns.drop(np.where(numerical_columns['Type']!='Photo')[0])
un_statuses=uncounted_columns.drop(np.where(uncounted_columns['Type']!='Status')[0])
statuses=numerical_columns.drop(np.where(numerical_columns['Type']!='Status')[0])
un_linkes=uncounted_columns.drop(np.where(uncounted_columns['Type']!='Link')[0])
linkes=numerical_columns.drop(np.where(numerical_columns['Type']!='Link')[0])
un_videos=uncounted_columns.drop(np.where(uncounted_columns['Type']!='Video')[0])
videos=numerical_columns.drop(np.where(numerical_columns['Type']!='Video')[0])

un_arr=[un_photos, un_statuses, un_linkes, un_videos]
arr=[photos, statuses, linkes, videos]
```

```python
for i in un_arr:
    print(i['Type'].unique())
    display(HTML(tabulate.tabulate(static_uncounted(i), tablefmt='html', headers=uncounted_columns.columns.values[1:])))
```

```python
for i in un_arr:
    print(i['Type'].unique())
    display(HTML(tabulate.tabulate(i.mode(), tablefmt='html', headers=uncounted_columns.columns.values[1:])))
```

```python
for i in arr:
    print(i['Type'].unique())
    display(HTML(tabulate.tabulate(static_counted(i), tablefmt='html', headers=numerical_columns.columns.values[1:])))
```
